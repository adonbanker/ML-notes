```python
import pandas as pd
import numpy as np
%pip install matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\Adon Banker\Downloads\canada_per_capita_income.csv")
```

```python
df
```

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Per Capita Income in Canada Over Years')
plt.show()
```

```python
r = LinearRegression()
r.fit(df[["year"]], df["per capita income (US$)"])  # you are training the model here using the data points
```

```python
r.predict([[2020]])  # you are predicting the per capita income for the year 2020
```

```python
r.predict([[2025]])  # you are predicting the per capita income for the year 2025
```

```python
r.coef_  # this is the slope of the line
```

```python
r.intercept_  # this is the y-intercept of the line
```
